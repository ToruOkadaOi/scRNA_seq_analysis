```python
an_object = ad.io.read_mtx("/home/rstudio/run070/run070-nsclc-3_RSEC_MolsPerCell_MEX/matrix.mtx.gz")
```

```python
scanpy_object = sc.read_mtx(
    "/home/rstudio/run070/run070-nsclc-3_RSEC_MolsPerCell_MEX/matrix.mtx.gz"
)
```

```python
scanpy_object.obs_names_make_unique()
#print(scanpy_object["sample"].value_counts())
scanpy_object
```

```python
# mitochondrial genes, "MT-" for human, "Mt-" for mouse
scanpy_object.var["mt"] = scanpy_object.var_names.str.startswith("MT-")
# ribosomal genes
scanpy_object.var["ribo"] = scanpy_object.var_names.str.startswith(("RPS", "RPL"))
# hemoglobin genes
scanpy_object.var["hb"] = scanpy_object.var_names.str.contains("^HB[^(P)]")
```

```python
#Calculating qc
sc.pp.calculate_qc_metrics(
    scanpy_object, qc_vars=["mt", "ribo", "hb"], inplace=True, log1p=True
)
```

```python
#Violin Plot
sc.pl.violin(
    scanpy_object,
    ["n_genes_by_counts", "total_counts", "pct_counts_mt"],
    jitter=0.4,
    multi_panel=True,
)
```

```python
#Scatter plot
sc.pl.scatter(scanpy_object, "total_counts", "n_genes_by_counts", color="pct_counts_mt")
```

```python
# filter cells with less than 100 genes expressed and genes that are detected in less than 3 cells
sc.pp.filter_cells(scanpy_object, min_genes=100)
sc.pp.filter_genes(scanpy_object, min_cells=3)
```

```python
#doublet detection - Scrublet method
sc.pp.scrublet(scanpy_object)
```

```python
# Saving count data
scanpy_object.layers["counts"] = scanpy_object.X.copy()
```

```python
#Norm.n
##count depth scaling with subsequent log plus one (log1p) transformation
### Normalizing to median total counts
sc.pp.normalize_total(scanpy_object)
### Logarithmize the data
sc.pp.log1p(scanpy_object)
```

```python
#Feature selection # batch_key if there are multiple sampels
sc.pp.highly_variable_genes(scanpy_object, n_top_genes=2000)
```

```python
sc.pl.highly_variable_genes(scanpy_object)
```

```python
#Dimensionality red.n
sc.tl.pca(scanpy_object)
```

```python
#pca
sc.pl.pca_variance_ratio(scanpy_object, n_pcs=50, log=True)
```

```python
#plotting pcs
sc.pl.pca(
    scanpy_object,
    color=["pct_counts_mt"], # Anything else to add together here??
    dimensions=[(0, 1), (2, 3), (0, 1), (2, 3)],
    ncols=2,
    size=2,
)
```

```python
#Nearest neighbour
sc.pp.neighbors(scanpy_object)
```

```python
#UMAP embedding
sc.tl.umap(scanpy_object)
```

```python
#grouping by gene counts ## later will group by cluster asfter running the clustering process
sc.pl.umap(scanpy_object, color="n_genes_by_counts", size=2)
```

```python
# Clustering
## Using the igraph implementation and a fixed number of iterations can be significantly faster, especially for larger datasets
sc.tl.leiden(scanpy_object, flavor="igraph", n_iterations=2)
```

```python
sc.pl.umap(scanpy_object, color=["leiden"])
```

```python
# qc using umap-plots
sc.pl.umap(
    scanpy_object,
    color=["leiden", "predicted_doublet", "doublet_score"],
    # increase horizontal space between panels
    wspace=0.6,
    size=3,-
)
```

```python
sc.pl.umap(
    scanpy_object,
    color=["leiden", "log1p_total_counts", "pct_counts_mt", "log1p_n_genes_by_counts"],
    wspace=0.5,
    ncols=2,
)
```

```python
# resolution control - clustering
for res in [0.02, 0.5, 2.0]:
    sc.tl.leiden(
        scanpy_object, key_added=f"leiden_res_{res:4.2f}", resolution=res, flavor="igraph"
    )
```

```python
sc.pl.umap(
    scanpy_object,
    color=["leiden_res_0.02", "leiden_res_0.50", "leiden_res_2.00"],
    legend_loc="on data",
)
```
